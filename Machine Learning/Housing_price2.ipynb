# Treatment of outliers using Imputing or Capping

```python
q1=housing["Sale Price"].quantile(0.25)
q3=housing["Sale Price"].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
upper_limit=q3 + 1.5*iqr
lower_limit=q1 - 1.5*iqr
upper_limit,lower_limit
```

```python
def limit_imputer(value):
    if value> upper_limit:
        return upper_limit
    elif value< lower_limit:
        return lower_limit
    else:
        return value
```

```python
housing["Sale Price"]=housing["Sale Price"].apply(limit_imputer)
```

```python
housing["Sale Price"].describe()
```

```python
housing_clean1=housing
```

```python
housing_clean1.dropna(inplace=True,axis=0,subset=["Sale Price"])
```

```python
housing_clean1.info()
```

```python
plt.hist(housing_clean1["Sale Price"],bins=10,color="green")
plt.xlabel("Intervals")
plt.ylabel("Selling Price")
plt.title("Histogram of deliing Price")
plt.show()
```
