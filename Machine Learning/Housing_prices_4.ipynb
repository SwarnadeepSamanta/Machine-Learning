```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
housing_reformed=housing_reformed.rename(columns={"Sale Price":"Sale_Price","Condition of the House":"Condition_of_the_House","Waterfront View":"Waterfront_View","Ever Renovated":"Ever_Renovated"})
```

```python
housing_reformed.head()
```

```python
mod=ols("Sale_Price ~ Condition_of_the_House",data=housing_reformed).fit()
```

```python
Anova_Table=sm.stats.anova_lm(mod,typ=2)
```

```python
print(Anova_Table)
```

```python
mod=ols("Sale_Price ~ Waterfront_View",data=housing_reformed).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
mod=ols("Sale_Price ~ Ever_Renovated",data=housing_reformed).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
mod=ols("Sale_Price ~ Zipcode",data=housing_reformed).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
housing_reformed=pd.get_dummies(housing_reformed,columns=["Condition_of_the_House","Ever_Renovated","Waterfront_View"],drop_first=True)
```

# Binning

```python
Zip_Table=housing_reformed.groupby("Zipcode").agg({"Sale_Price":"mean"}).sort_values("Sale_Price",ascending=True)
```

```python
Zip_Table.head()
```

```python
Zip_Table["Zipcode_Group"]=pd.cut(Zip_Table["Sale_Price"],bins=10,labels=["Zipcode_Group_0","Zipcode_Group_1","Zipcode_Group_2","Zipcode_Group_3","Zipcode_Group_4","Zipcode_Group_5","Zipcode_Group_6","Zipcode_Group_7","Zipcode_Group_8","Zipcode_Group_9"],include_lowest=True)
```

```python
Zip_Table=Zip_Table.drop(columns="Sale_Price")
```

```python
housing_reformed=pd.merge(housing_reformed,Zip_Table,left_on="Zipcode",how="left",right_index=True)
```

```python
housing_reformed=housing_reformed.drop(columns="Zipcode")
```

```python
housing_reformed.head()
```

```python
housing_reformed=pd.get_dummies(housing_reformed,columns=["Zipcode_Group"],drop_first=True)
```
