```python
housing['Condition of the House']=housing['Condition of the House'].map({'Good':'1','Excellent':'3','Bad':'0','Fair':'1','Okay':'0'})
housing['Condition of the House'].unique()
```

```python
#Extracting year from date of selling
#Normal way
year=[]
for i in range(len(housing["Date House was Sold"])):
    k=housing["Date House was Sold"][i].split()[-1]
    year.append(k)
    
housing["year_sold"]=year
housing["year_sold"].head()
```

```python
#Extracting year from date of selling
#Using Pandas library
def year(value):
    return value.split()[-1]

housing["year_sold"]=housing["Date House was Sold"].map(year)
housing["year_sold"].head()
```

```python
#Find out if the house is a luxuary house or not?
#if it has waterfront view,comndition of the house is in good or excellent satate and overall grade geater than equals to 8.
#if ant 2 or more conditions are met then it is luxuary house else not.
def luxury_home(row):
    count=0
    if row[0]=='Yes':
        count+=1
    if row[1] in ["Good","Excellent"]:
        count+=1
    if row[2]>=8:
        count+=1
    if count>=2:
        return "Yes"
    else:
        return "NO"


housing["luxury_home"]=housing[['Waterfront View','Condition of the House','Overall Grade']].apply(luxury_home,axis=1)
housing["luxury_home"].unique()
#housing["luxury_home"].head(100)
```
