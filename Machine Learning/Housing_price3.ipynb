```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",500)
pd.set_option("display.max_rows",10000000)
```

```python
housing=pd.read_csv("New folder/Raw_Housing_Prices3.csv")
```

```python
housing.info()
```

```python
numerical_columns=['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='median')
housing[numerical_columns]=imputer.fit_transform(housing[numerical_columns])
```

```python
housing.info()
```

```python
housing['Zipcode'].shape
```

```python
column=housing['Zipcode'].values.reshape(-1,1)
column.shape
```

```python
column=housing["Zipcode"].values.reshape(-1,1)
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
housing['Zipcode']=imputer.fit_transform(column)
```

```python
housing.info()
```

```python
housing['Zipcode']=housing['Zipcode'].astype(object)
```

```python
housing.info()
```

```python
housing["No of Times Visited"].unique()
```

```python
mapping={'None' : "0", 'Thrice' : "3", 'Four' : "4", 'Twice' : "2", 'Once' : "1"}
housing["No of Times Visited"]=housing["No of Times Visited"].map(mapping)
```

```python
housing["No of Times Visited"].unique()
```

```python
housing["Ever Renovated"]=np.where(housing["Renovated Year"]==0,"No","Yes")
```

```python
housing.head()
```

```python
housing["Purchase Year"]=pd.DatetimeIndex(housing["Date House was Sold"]).year
```

```python
housing["Year Since Renovation"]=np.where(housing["Ever Renovated"]=="Yes",abs(housing["Purchase Year"]-housing["Renovated Year"]),0)
```

```python
housing.head()
```

```python
housing.drop(columns=["Purchase Year","Date House was Sold","Renovated Year"],inplace=True)
```

```python
housing.head()
```

```python
housing.info()
```

```python
housing.to_csv('Transformed_Housing_Data_me.csv',index=False)
```
